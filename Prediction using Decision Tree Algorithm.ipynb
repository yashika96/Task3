# Task 3: Prediction using Decision Tree Algorithm

We will create a decision tree classifier for the 'Iris' dataset and visualise it graphically. Classfication trees are relevant here since our target variable is categorical in nature.

### Importing Libraries

```python
# Importing all the required libraries

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn

from sklearn import datasets
from sklearn.tree import DecisionTreeClassifier 
from sklearn.model_selection import train_test_split 
from sklearn import metrics
from sklearn.metrics import confusion_matrix, classification_report

from sklearn.tree import export_graphviz 
from IPython.display import Image  
import pydotplus
```

### Loading Data

```python
# Loading the data

iris = datasets.load_iris()
X = iris.data
y = iris.target
X[0:10]
```

### Splitting the Data

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1)

print("Training Complete")
```

### Building Decision Tree Model

```python
clf = DecisionTreeClassifier()
clf = clf.fit(X_train,y_train)

print('Decision Tree Classifer Created')
```

### Predictions

```python
y_pred = clf.predict(X_test)
print("Predicted values:", y_pred)
```

```python
df = pd.DataFrame({'Actual': y_test, 'Predicted': y_pred})
df.head()
```

### Evaluating the Model

```python
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))   
```

```python
print("Report : ", classification_report(y_test, y_pred))
```

We got classification rate of 95.56% which is considered as good accuracy score.Classification report suggests high recall and high precision which describes model's completeness and relevancy respectively.

### Visualizing Decision Trees

```python
from sklearn.externals.six import StringIO
df=pd.DataFrame(iris.data, columns=iris.feature_names)

dot_data = StringIO()
export_graphviz(clf, out_file=dot_data,  
                filled=True, rounded=True,
                special_characters=True,feature_names = iris.feature_names )
graph = pydotplus.graph_from_dot_data(dot_data.getvalue())  
Image(graph.create_png())
```
